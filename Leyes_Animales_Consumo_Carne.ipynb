## Visualización de Datos

```python
df1 = df.groupby(["Support for the Universal Declaration of Animal Welfare","Fur-Farming Ban"])["Meat Consumption per Capita in Kilograms"].mean().reset_index()
df1
```

```python
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots()
ax.bar(df1["Support for the Universal Declaration of Animal Welfare"],df1["Meat Consumption per Capita in Kilograms"])
ax.set_xlabel("Support animals")
ax.set_ylabel("Average meat consumption per kg")
ax.set_title("Support for the Universal Declaration of Animal Welfare vs Meat Consumption per Capita in Kilograms",fontsize = 8)
plt.show()
```

Notamos que el consumo de carne aumenta aunque exista apoyo por el bienestar animal.

```python
fig, ax = plt.subplots()
ax.bar(df1["Fur-Farming Ban"],df1["Meat Consumption per Capita in Kilograms"])
ax.set_xlabel("Fur farming ban",fontsize = 10)
ax.set_ylabel("Average meat consumption per kg",fontsize = 11)
ax.set_title("Fur farming ban vs Meat Consumption per Capita in Kilograms")
plt.show()
```

```python
values = df["Fur-Farming Ban"].value_counts()
values
```

```python
etiquetas=["No", "Partial","Yes"]
explotar=[0.02,0.03,0.03]
fig, ax = plt.subplots()
ax.set_title("Fur farming ban")
ax.pie(values,labels=etiquetas,autopct="%.1f%%",explode=explotar,startangle=98)
plt.show()
```
